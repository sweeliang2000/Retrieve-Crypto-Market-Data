```python
import ccxt
from datetime import datetime
import pandas as pd
```

```python
exchange_id = "kucoin"
```

```python
exchange_class = getattr(ccxt, exchange_id)
```

```python
exchange = exchange_class({
    'apiKey': '61fb73124cf8fb0001417935',
    'secret': '767cf515-0566-4992-87e9-ecad34f0ed90',
})
```

```python
exchange.load_markets()
```

```python
def retrieve_date(coin,timeframe):
    btc = pd.DataFrame(exchange.fetch_ohlcv(coin,timeframe=timeframe))
    btc = btc.rename(columns={0: "time", 1: "Open",2:"High",3:"Low",4:'Close',5:'Volume'})
    newlist = []
    for i in btc['time']:
        newlist.append(datetime.fromtimestamp(i/1000.0))
    btc['date']= newlist
    btc= btc.set_index("date")
    btc = btc.drop('time', 1)
    return btc
```

```python
btc = retrieve_date("BTC/USDT",'1h')
```
